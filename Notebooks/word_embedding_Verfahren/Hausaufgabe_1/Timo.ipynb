# Code Beispiel (tf-idf Modell)

https://towardsdatascience.com/tf-idf-explained-and-python-sklearn-implementation-b020c5e83275

TF-IDF steht für Term Frequency - Inverse Document Frequency und ist eine Statistik, die darauf abzielt, besser zu definieren, wie wichtig ein Wort für ein Dokument ist, wobei auch die Beziehung zu anderen Dokumenten aus demselben Korpus berücksichtigt wird.

Dazu wird untersucht, wie oft ein Wort in einem Dokument vorkommt, wobei auch darauf geachtet wird, wie oft das gleiche Wort in anderen Dokumenten im Korpus vorkommt.

&rarr; ein Wort, das häufig in einem Dokument vorkommt, hat eine höhere Relevanz für dieses Dokument, d. h., es besteht eine höhere Wahrscheinlichkeit, dass das Dokument von diesem bestimmten Wort handelt oder sich darauf bezieht.

&rarr; ein Wort, das häufig in mehreren Dokumenten vorkommt, kann uns daran hindern, das richtige Dokument in einer Sammlung zu finden; das Wort ist entweder für alle Dokumente oder für keines relevant. So oder so wird es uns nicht helfen, ein einzelnes Dokument oder eine kleine Teilmenge von Dokumenten aus der gesamten Menge herauszufiltern.

&rarr; Der TF-IDF ist ein Score, der auf jedes Wort in jedem Dokument in unserem Datensatz angewendet wird. Für jedes Wort steigt der TF-IDF-Wert mit jedem Auftreten des Wortes in einem Dokument, wird aber mit jedem Auftreten in anderen Dokumenten allmählich verringert.

# PMI Model

https://www.kaggle.com/gabrielaltay/word-vectors-from-pmi-matrix

Erstellund von einfachen Wortvektoren, indem eine Singulärwertzerlegung auf eine Wort-Wort-Matrix mit punktweiser gegenseitiger Information angewendet wird. Es gibt viele andere Möglichkeiten, Wortvektoren zu erzeugen, aber die Matrixzerlegung ist eine der einfachsten.

```python
from collections import Counter
import itertools

import nltk
from nltk.corpus import stopwords
import numpy as np
import pandas as pd
from scipy import sparse
from scipy.sparse import linalg 
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("../../Datensatz/df_original.csv")
df
```

## Minimal Preprocessing

Es wird eine Wort-Wort-Co-Occurrence-Matrix aus den Texten erstellt. Wir definieren zwei Wörter als "co-occurring", wenn sie im gleichen Text vorkommen. Mit dieser Definition sind alleinstehende Wörter uninteressant. Wir entfernen sie ebenso wie eine Reihe von englischen Stoppwörtern.

```python
headlines = df['clean_text'].tolist()
# Stoppwörter entfernen
stopwords_set = set(stopwords.words('german'))
headlines = [
    [tok for tok in headline.split() if tok not in stopwords_set] for headline in headlines
]
# Entfernen alleinstehender Wörter
headlines = [hl for hl in headlines if len(hl) > 1]
# Ergebis anzeigen
headlines[0:20]
```

## Unigrams

Berechnung des Unigramm-Vokabulars. Der folgende Code weist jedem Token eine eindeutige ID zu, speichert diese Zuordnung in zwei Wörterbüchern (tok2indx und indx2tok) und zählt, wie oft jedes Token im Korpus erscheint.

```python
unigram_counts = Counter()
for ii, headline in enumerate(headlines):
    if ii % 200000 == 0:
        print(f'finished {ii/len(headlines):.2%} of headlines')
    for token in headline:
        unigram_counts[token] += 1

tok2indx = {tok: indx for indx, tok in enumerate(unigram_counts.keys())}
indx2tok = {indx: tok for tok,indx in tok2indx.items()}
print('done')
print('vocabulary size: {}'.format(len(unigram_counts)))
print('most common: {}'.format(unigram_counts.most_common(10)))
```

## Skipgrams

Berechnung eines Skipgramm-Vokabulas. Es wird für jedes Wort in einem Text (das Fokuswort) eine Schleife durchlaufen und dann Skipgramme gebildet, indem wir back_window-Wörter hinter und front_window-Wörter vor dem Fokuswort (die Kontextwörter) untersuchen.

```python
# Die Token-Vokabular-Indizes werden im Skipgramm-Zähler gespeichert

back_window = 2
front_window = 2
skipgram_counts = Counter()
for iheadline, headline in enumerate(headlines):
    tokens = [tok2indx[tok] for tok in headline]
    for ii_word, word in enumerate(tokens):
        ii_context_min = max(0, ii_word - back_window)
        ii_context_max = min(len(headline) - 1, ii_word + front_window)
        ii_contexts = [
            ii for ii in range(ii_context_min, ii_context_max + 1) 
            if ii != ii_word]
        for ii_context in ii_contexts:
            skipgram = (tokens[ii_word], tokens[ii_context])
            skipgram_counts[skipgram] += 1    
    if iheadline % 200000 == 0:
        print(f'finished {iheadline/len(headlines):.2%} of headlines')
        
print('done')
print('number of skipgrams: {}'.format(len(skipgram_counts)))
most_common = [
    (indx2tok[sg[0][0]], indx2tok[sg[0][1]], sg[1]) 
    for sg in skipgram_counts.most_common(10)]
print('most common: {}'.format(most_common))
```

## Word-Word Count Matrix

Die ersten Wortvektoren werden aus einer Wort-Wortanzahl-Matrix stammen. Diese Matrix ist symmetrisch, so dass man (äquivalent) die Wortvektoren als die Zeilen oder Spalten betrachten kann. Es wird jedoch versucht, so zu kodieren, als ob die Zeilen Wortvektoren und die Spalten Kontextvektoren sind.

```python
row_indxs = []
col_indxs = []
dat_values = []
ii = 0
for (tok1, tok2), sg_count in skipgram_counts.items():
    ii += 1
    if ii % 1000000 == 0:
        print(f'finished {ii/len(skipgram_counts):.2%} of skipgrams')    
    row_indxs.append(tok1)
    col_indxs.append(tok2)
    dat_values.append(sg_count)
wwcnt_mat = sparse.csr_matrix((dat_values, (row_indxs, col_indxs)))
print('done')
```

## Word Similarity with Sparse Count Matrices

```python
def ww_sim(word, mat, topn=10):
    """Berechnen der Topn der ähnlichsten Wörter zum Wort"""
    indx = tok2indx[word]
    if isinstance(mat, sparse.csr_matrix):
        v1 = mat.getrow(indx)
    else:
        v1 = mat[indx:indx+1, :]
    sims = cosine_similarity(mat, v1).flatten()
    sindxs = np.argsort(-sims)
    sim_word_scores = [(indx2tok[sindx], sims[sindx]) for sindx in sindxs[0:topn]]
    return sim_word_scores
    
```

## Pointwise Mutual Information Matrices

Die punktweise gegenseitige Information (PMI) für ein Paar (Wort, Kontext) ist definiert als die Wahrscheinlichkeit ihres gemeinsamen Auftretens geteilt durch die Wahrscheinlichkeiten ihres individuellen Auftretens.

```python
num_skipgrams = wwcnt_mat.sum()
assert(sum(skipgram_counts.values())==num_skipgrams)

# Erstellung von sparce matrices
row_indxs = []
col_indxs = []

pmi_dat_values = []    # punktweise gegenseitige Information
ppmi_dat_values = []   # positive punktweise gegenseitige Information
spmi_dat_values = []   # geglättete punktweise gegenseitige Information
sppmi_dat_values = []  # geglättete positive punktweise gegenseitige Information

# Mehrwegmengen

# sum_over_rows[ii] = sum_over_words[ii] = wwcnt_mat.getcol(ii).sum()
sum_over_words = np.array(wwcnt_mat.sum(axis=0)).flatten()
# sum_over_cols[ii] = sum_over_contexts[ii] = wwcnt_mat.getrow(ii).sum()
sum_over_contexts = np.array(wwcnt_mat.sum(axis=1)).flatten()

# Glätten
alpha = 0.75
sum_over_words_alpha = sum_over_words**alpha
nca_denom = np.sum(sum_over_words_alpha)



ii = 0
for (tok_word, tok_context), sg_count in skipgram_counts.items():

    ii += 1
    if ii % 1000000 == 0:
        print(f'finished {ii/len(skipgram_counts):.2%} of skipgrams')

    
    nwc = sg_count
    Pwc = nwc / num_skipgrams
    nw = sum_over_contexts[tok_word]
    Pw = nw / num_skipgrams
    nc = sum_over_words[tok_context]
    Pc = nc / num_skipgrams
    
    nca = sum_over_words_alpha[tok_context]
    Pca = nca / nca_denom
    
    pmi = np.log2(Pwc/(Pw*Pc))   
    ppmi = max(pmi, 0)
    spmi = np.log2(Pwc/(Pw*Pca))
    sppmi = max(spmi, 0)
    
    row_indxs.append(tok_word)
    col_indxs.append(tok_context)
    pmi_dat_values.append(pmi)
    ppmi_dat_values.append(ppmi)
    spmi_dat_values.append(spmi)
    sppmi_dat_values.append(sppmi)
        
pmi_mat = sparse.csr_matrix((pmi_dat_values, (row_indxs, col_indxs)))
ppmi_mat = sparse.csr_matrix((ppmi_dat_values, (row_indxs, col_indxs)))
spmi_mat = sparse.csr_matrix((spmi_dat_values, (row_indxs, col_indxs)))
sppmi_mat = sparse.csr_matrix((sppmi_dat_values, (row_indxs, col_indxs)))

print('done')
```

## Singular Value Decomposition

Mit den PMI- und PPMI-Matrizen kann nun durch eine Singulärwertzerlegung aus den bisher spärlichen Wortvektoren dichte Wortvektoren erzeugt werden.

```python
pmi_use = ppmi_mat
embedding_size = 50
uu, ss, vv = linalg.svds(pmi_use, embedding_size) 
```

```python
def word_sim_report(word, sim_mat):
    sim_word_scores = ww_sim(word, word_vecs)
    for sim_word, sim_score in sim_word_scores:
        print(sim_word, sim_score)
        word_headlines = [hl for hl in headlines if sim_word in hl and word in hl][0:5]
        for headline in word_headlines:
            print(f'    {headline}')
```
