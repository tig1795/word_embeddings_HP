# Graphische Modellierung von Textähnlichkeiten: Dendrogramm

### "Harry Potter"-Korpus

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Einlesen des "Harry Potter"-Datensatzes in dem Dataframe "df" 

```python
###nach den Notebooks "Sophia.ipynb" und "Timo.ipynb"
df = pd.read_csv("/home/iuser/DH/DH_2020_2021/word_embeddings_HP/Datensatz/df_original.csv")
df
```

#### Document-Term-Matrix →  tf-idf-Matrix → Dendrogramm

```python
###Code nach dem Notebook "Timo.ipynb"
#Erstellen einer tf-idf-Matrix (gewichtete Kookkurenz) aus der Document-Term-Matrix (reine Termhäufigkeit) 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords

#Definition der zu beseitigenden Stoppwörter
german_stop_words = stopwords.words('german')

#Vektorisierung der Rohdokumenten und Umsetzung in eine Matrix von TF-IDF-Merkmalen
vectorizer = CountVectorizer(stop_words = german_stop_words)
X = vectorizer.fit_transform(df.clean_text)

numpy_x = X.toarray()

df10 = pd.DataFrame(data=numpy_x, index=df.Titel, 
                   columns=vectorizer.get_feature_names())
df10
```

```python
###Quellcode: http://brandonrose.org/clustering

#Als Basis für die Erstellung des Dendrogramms wird die tf-idf-Matrix benutzt
#Die Cosinus-Ähnlichkeit wird hier als Distanzmaß verwendet, aber der Euklidische-(default), Mahnanttan- und andere
#Abstandsmaße sind auch möglich.
from sklearn.metrics.pairwise import cosine_similarity
dist_df = 1 - cosine_similarity(df10)   #"dist_df" entspricht nur positive Abstandswerte
dist_df
```

```python
#Die linkage_matrix_df, durch die untenstehende Funktion, nutzt Distanzmaß und Methode, um den Abstand zwischen 
#den Dokumenten des Korpus zu bestimmen. Ähnliche Dokumente stellen Cluster dar.
#Methoden bzw. Algorithmen können z.B. 'single', 'complete', 'average', 'ward' sein.
#Bei jeder Iteration, werden die zwei Cluster mit dem geringsten Abstand miteinander gemischt ("merged").

from scipy.cluster.hierarchy import ward, dendrogram
linkage_matrix_df = ward(dist_df) #Anwendung der ward-Clustering-Methode auf die vorberechneten Abstandswerte "dist_df"
linkage_matrix_df
```

```python
#Eigenschaften und Ausgeben des Dendrogramms
fig, ax = plt.subplots(figsize=(10,10))
ax = dendrogram(linkage_matrix_df, orientation="right", labels=df10.index)

plt.title('Harry Potter - Hierarchical Clustering Dendrogram')
plt.ylabel('Documents')
plt.xlabel('Distance')
plt.tight_layout()
```
